```python
import matplotlib.pyplot as plt

# Data
server_elements = [100]*3 + [1000]*3 + [10000]*3
server_projections = [10, 20, 30] * 3

# Raw values in milliseconds
server_total = [
    80321, 80224, 97708,
    913388, 913499, 907532,
    11967835, 11962672, 12008961
]
server_exp = [
    59820, 59816, 59851,
    893017, 893100, 887074,
    11947016, 11941324, 11987392
]
server_dhf = [
    20487, 20382, 20749,
    20357, 20377, 20411,
    20713, 21226, 21434
]

# Convert times
server_total_min = [x / 60000 for x in server_total]
server_exp_min = [x / 60000 for x in server_exp]
server_dhf_sec = [x / 1000 for x in server_dhf]

# Organize by element count
server_data = {
    100:   {'proj': [10, 20, 30], 'total': server_total_min[0:3], 'exp': server_exp_min[0:3], 'dhf': server_dhf_sec[0:3]},
    1000:  {'proj': [10, 20, 30], 'total': server_total_min[3:6], 'exp': server_exp_min[3:6], 'dhf': server_dhf_sec[3:6]},
    10000: {'proj': [10, 20, 30], 'total': server_total_min[6:9], 'exp': server_exp_min[6:9], 'dhf': server_dhf_sec[6:9]},
}

# Plotting
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 10), sharex=True)

# Total compute time (minutes)
for n in server_data:
    ax1.plot(server_data[n]['proj'], server_data[n]['total'], marker='o', label=f'{n} elements')
ax1.set_title("Server Total Compute Time")
ax1.set_ylabel("Time (minutes)")
ax1.legend()
ax1.grid(True)
ax1.set_xticks([10, 20, 30])

# Exponentiation time (minutes)
for n in server_data:
    ax2.plot(server_data[n]['proj'], server_data[n]['exp'], marker='s', label=f'{n} elements')
ax2.set_title("Server Exponentiation Time")
ax2.set_ylabel("Time (minutes)")
ax2.legend()
ax2.grid(True)

# DHF time (seconds)
for n in server_data:
    ax3.plot(server_data[n]['proj'], server_data[n]['dhf'], marker='^', label=f'{n} elements')
ax3.set_title("Server DHF Time")
ax3.set_ylabel("Time (seconds)")
ax3.set_xlabel("Number of Projections")
ax3.legend()
ax3.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Updated full client data
client_elements = [100]*3 + [1000]*3 + [10000]*3
client_projections = [10, 20, 30] * 3
client_total_ms = [
    32609, 65157, 97708,       # 100
    32424, 64869, 97472,      # 1000
    32579, 65306, 97629        # 10000
]
client_dhf_ms = [
    205, 205, 205,
    190, 190, 198,
    191, 191, 193
]

# Convert total compute time to seconds
client_total_sec = [x / 1000 for x in client_total_ms]

# Organize by element size
client_data = {
    100:   {'proj': [], 'total': [], 'dhf': []},
    1000:  {'proj': [], 'total': [], 'dhf': []},
    10000: {'proj': [], 'total': [], 'dhf': []}
}

for i in range(len(client_elements)):
    n = client_elements[i]
    p = client_projections[i]
    total = client_total_sec[i]
    dhf = client_dhf_ms[i]

    client_data[n]['proj'].append(p)
    client_data[n]['total'].append(total)
    client_data[n]['dhf'].append(dhf)

# Plotting
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

# Total compute time (seconds)
for n in client_data:
    ax1.plot(client_data[n]['proj'], client_data[n]['total'], marker='o', label=f'{n} elements')
ax1.set_title("Client Compute Time")
ax1.set_ylabel("Time (seconds)")
ax1.legend()
ax1.grid(True)
ax1.set_xticks([10, 20, 30])


# DHF time (ms)
for n in client_data:
    ax2.plot(client_data[n]['proj'], client_data[n]['dhf'], marker='^', label=f'{n} elements')
ax2.set_title("Client DHF Time")
ax2.set_ylabel("Time (milliseconds)")
ax2.set_xlabel("Number of Projections")
ax2.legend()
ax2.grid(True)


plt.tight_layout()
plt.show()
```
